```python
import matplotlib.pyplot as plt
import numpy as np
 
def funkcja(x):
    return x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)


def wykres(x):
    plt.figure(figsize=(8,5))
    plt.plot(x, funkcja(x))
    plt.xlabel("argumenty")
    plt.ylabel("wartosci")
    plt.legend("f")
    plt.grid()
    plt.title("Wykres funkcji")
    plt.show()

wykres(x1)
wykres(x2)
wykres(x3)
```
