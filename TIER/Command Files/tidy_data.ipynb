```python
import pandas as pd
import numpy as np

df = pd.read_csv("../Original Data/tb.csv")
df
```

**UsuniÄ™cie NaN**

```python
df = df.dropna()
df = df.drop(columns=['new_sp'])
```

**Nazwanie kolumn**

```python
df.rename(columns={'new_sp_m04':'m04', 
                   'new_sp_m514':'m514',      
                   'new_sp_m014':'m014',
                   'new_sp_m1524':'m1524',
                   'new_sp_m2534':'m2534',
                   'new_sp_m3544':'m3522',
                   'new_sp_m5564':'m5564',      
                   'new_sp_f04':'f04',
                   'new_sp_f514':'f514',
                   'new_sp_f014':'f014',
                   'new_sp_f1524':'f1524',
                   'new_sp_f2534':'f2534',      
                   'new_sp_f3544':'f3544',
                   'new_sp_f4554':'f4554',
                   'new_sp_f5564':'f5564',
                   'new_sp_f65':'f65', 
                   'new_sp_m65':'m65',
                   'new_sp_fu':'fu', 
                   'new_sp_mu':'mu', 
                   'new_sp_m4554':'m4554'}, inplace=True)
df
```

**Zamiana formatu na int**

```python
lista_kolumn = list(df.columns)
lista_kolumn_2 = lista_kolumn[-20:]
df[lista_kolumn_2] = df[lista_kolumn_2].astype(int)
df
```

```python
df.describe()
```

**UporzÄ…dkowanie**

```python
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[2:],var_name='column', value_name='cases')
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[0]
df['age'] = df['column'].str[1:].map({
    '04': '0 - 4',
    '514': '5 - 14',
    '1524': '15 - 24',
    '2534': '25 - 34',
    '3544': '35 - 44',
    '4554': '45 - 54',
    '5564': '55 - 64',
    '65': '65+',
    'u' : 'unknown',
    '' : 'unknown',
  
})
df = df[['iso2', 'year', 'sex', 'age', 'cases']]
df.to_csv('../Analysis Data/tb_tidy.csv')
df
```

**NajwiÄ™cej przypadkÃ³w**  


```python
high = df.sort_values("cases", ascending = False)
display(high)
```

**Tylko mÄ™Å¼czyÅºni**  


```python
df_male = df.groupby(df.sex).get_group('m')
df_male_final = pd.DataFrame(df_male.groupby(['age', 'cases']).size().groupby(level=0).max()).reset_index()
df_male_final
```
